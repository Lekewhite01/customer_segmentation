```python
import pandas as pd
import numpy as np
from make_plots import barplot, histogram
```

```python
# Read labelled data into pandas dataframe
df = pd.read_csv("./output/labelled_data.csv")

# Convert `year` to string in the main dataframe
df["order_purchase_year"] = df["order_purchase_year"].astype(str)
```

```python
df.head()
```

## Cluster 0

```python
# Subset the dataframe to just `cluster 0` profiles
cluster_0_df = df[df["clusters"]==0]
cluster_0_df.head()
```

```python
# Compute number of customers in this cluster
customers = cluster_0_df["customer_unique_id"].nunique()
print("Number of customers in the Cluster 0:", customers)
```

```python
# Group by both 'order_year' and 'order_month' and count the number of orders in each month
monthly_orders = cluster_0_df.groupby(['order_purchase_year', 'order_purchase_month']).size().reset_index(name='num_orders')

# Group by 'order_year' and calculate the average number of orders per month to two decimal places
average_monthly_orders = monthly_orders.groupby('order_purchase_year')['num_orders'].mean().reset_index(name='average_monthly_orders').round(2)

barplot(average_monthly_orders,
        x="order_purchase_year",
        y="average_monthly_orders",
        title="Average Orders per month",
        xaxis="Year", 
        yaxis="Average Monthly Orders")
```

```python
# Compute customer distribution by state
customers_per_state = cluster_0_df.groupby("customer_state").size().reset_index(name='num_customers').sort_values("num_customers", ascending=False)

# Plot the customer distrbution by state
barplot(df=customers_per_state,
        x="customer_state",
        y="num_customers",
        title="Customer distribution by State",
        xaxis="Year", 
        yaxis="Number of Customers")
```

```python
# Compute customer distribution by payment type
customers_per_channel = cluster_0_df.groupby("payment_type").size().reset_index(name='num_customers').sort_values("num_customers", ascending=False)

# Plot the customer distrbution by payment type
barplot(df=customers_per_channel,
        x="payment_type",
        y="num_customers",
        title="Customer distribution by Payment Type",
        xaxis="Payment Type", 
        yaxis="Number of Customers")
```

```python
# Compute customer distribution by product category
customers_per_product = cluster_0_df.groupby("product_category").size().reset_index(name='num_customers').sort_values("num_customers", ascending=False)

# Plot the customer distrbution by product category
barplot(df=customers_per_product[:10],
        x="product_category",
        y="num_customers",
        title="Customer distribution by Product Type",
        xaxis="Product Type", 
        yaxis="Number of Customers")
```

```python
# Plot Price Distribution
histogram(df=cluster_0_df,
          x="price", 
          title="Price Distribution for Cluster 0",
          xaxis="Price")
```

```python
# Plot Payment Distribution
histogram(df=cluster_0_df,
          x="payment_value", 
          title="Payment Value Distribution for Cluster 0",
          xaxis="Payment Value")
```

```python
# Plot Installment Distribution
histogram(df=cluster_0_df,
          x="payment_installments", 
          title="Payment Installments Distribution for Cluster 0",
          xaxis="Payment Installments")
```
